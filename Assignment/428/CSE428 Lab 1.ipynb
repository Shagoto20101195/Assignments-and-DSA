```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, img_as_float
```

```python
# Adding 2 to all values to the list using numpy
x = [2, 3, 4, 5]

x = np.array(x)
x = x + 2
print(x)
```

```python
# 2D array
x = [[1, 2, 3], [4,5,6]]
x = np.array(x)
print(x)
```

```python
# 3D array for images - just stack 2D arrays in another numpy array
```

```python
# x[row, col]
# x[row][col]
print(x[1, 1])
print(x[1][1])
```

```python
# x[start : stop : step] -> [x[start], x[start + step], x[start + step + step] ..... x[stop-1]]
# x[::-1] -> start = -1, stop = -len(x) - 1
```

```python
# Row 1 and everything
x = np.array([[1, 2, 3], [4, 5, 6]])

print(x[1, :])
```

```python
# Row 1 and everything
x = np.array([[1, 2, 3], [4, 5, 6]])

print(x[:,1])
```

```python
# 1 2 3
# 4 5 6
# Showing without 2 and 5
x = np.array([[1, 2, 3], [4, 5, 6]])
print(x[:, ::2])
```

```python
x = np.array([[1, 2, 3], [4, 5, 6]])
print(x)
```

```python
# Rotating column (flipping across y, where x changes only)
print(x[:, ::-1])
```

```python
print(x[::-1, ::-1])
```

```python
# Boolean indexing
# i.e selective index will be true and rest will be false
x = np.array([1, 2, -3, 4, 5])
index = [False, True, False, True, False]
print(x[index])
```

```python
# Instead of making index list manually, we can:
index = x > 0
print(index)
print(x[index])
print(x[x > 0])
```

```python
# Image - 2D i.e. grayscale 0 - pure black and 1 - pure white (in between - gray).
# Row = Height, Col = Width
# Here Row = Height = 4 and Col = Width = 5
x = np.random.rand(4, 5)
print(x)
print(x.shape)
```

```python
io.imshow(x)
plt.axis('off')
```

```python
# Image - 3D. Color image - R G B
# Dimensions - row, column, channel (rgb)
x = np.random.rand(4, 5, 3)
# All color channel
red = x[:, :, 0]
print(red)
green = x[:, :, 1]
print(green)
blue = x[:, :, 2]
print(blue)
```

```python
io.imshow(x)
plt.axis('off')
```

```python
# Important to note the color mixing
# e.g. 1st row last col cell
# r = 0.74, g = 0.84, b = 0.16
# multiply by 255 and match color with https://www.w3schools.com/colors/colors_picker.asp rgb(r, g, b)
# 0 to 1 means normalized
```
