```python
import random

def make_ans(l):
  temp = ""
  for i in l:
    temp += str(i)
  return temp

def get_sum(amount, l):
  sum = 0
  for i in range(len(amount)):
    if(l[i] == 1):
      sum += amount[i]
  return abs(sum)

def make_list(n):
  l = []
  for i in range(n):
    l.append(random.randint(0, 1))
  return l

def crossover(gen_list, n):
  index = None
  if(n == 1):
    return [0], [1]
  elif(n == 2):
    index = 1
  else:
    index = random.randint(1, n-2)

  l1 = random.randint(0, 5) # First list to be crossovered
  l2 = random.randint(0, 5) # Second list to be crossovered
  while(l1 == l2):
    l2 = random.randint(0, 5)
  
  new1 = gen_list[l1][:index] + gen_list[l2][index:]
  new2 = gen_list[l2][:index] + gen_list[l1][index:]
  return new1, new2

def mutation(l):
  index = random.randint(0, n-1)
  l[index] = 1 - l[index]
  return l

ans = -1
count = 0
amount = []
n = int(input())
for i in range(n):
  inp = input().split(" ")

  if(inp[0] == "l"):
    amount.append(int(inp[1]) * -1)
  else:
    amount.append(int(inp[1]))

while(ans == -1 and count < 10000):
  # Randomly generate
  # All zero not allowed
  gen_list = []
  for i in range(8):
    temp = make_list(n)
    while temp.count(0) == n:
      temp = make_list(n)
    gen_list.append(temp)
  
  # Calculating fitness
  # Big value = Less fit
  max_sum = -99999999999
  for i in gen_list:
    res = get_sum(amount, i)
    max_sum = max(max_sum, res)
    if(res == 0) and i.count(0) != n:
      ans = make_ans(i)
      break
  
  # If found after fitness check then break
  if(ans != -1):
    break

  # Else proceed to selection - removing two least fit
  for i in range(8):
    if(get_sum(amount, gen_list[i]) == max_sum):
      gen_list.pop(i)
      break

  max_sum = -99999999999
  for i in gen_list:
    res = get_sum(amount, i)
    max_sum = max(max_sum, res)

  for i in range(7):
    if(get_sum(amount, gen_list[i]) == max_sum):
      gen_list.pop(i)
      break
  
  # Crossover
  new1, new2 = crossover(gen_list, n)
  while (new1.count(0) == n or new2.count(0) == n) and n > 1:
    new1, new2 = crossover(gen_list, n)

  # After crossover, check sum
  if(get_sum(amount, new1) == 0) and new1.count(0) != n:
    ans = make_ans(new1)
    break
  if(get_sum(amount, new2) == 0) and new2.count(0) != n:
    ans = make_ans(new2)
    break
  
  # Mutation
  new1 = mutation(new1)
  new2 = mutation(new2)
  while new1.count(0) == n and n > 1:
    new1 = mutation(new1)
  while new2.count(0) == n and n > 1:
    new2 = mutation(new2)

  # After mutation, check sum
  if(get_sum(amount, new1) == 0) and new1.count(0) != n:
    ans = make_ans(new1)
    break
  if(get_sum(amount, new2) == 0) and new2.count(0) != n:
    ans = make_ans(new2)
    break
    
  count += 1

print(ans)
```
