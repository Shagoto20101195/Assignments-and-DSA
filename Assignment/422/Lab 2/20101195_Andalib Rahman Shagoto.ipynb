```python
import random
import math

# Left node = 2n+1
# Right node = 2n+2
def dfs(graph, alpha, beta, level, node_number):
  if(level == 2):
    left_node = graph[2*node_number + 1]
    right_node = graph[2*node_number + 2]

    if(left_node > right_node):
      graph[node_number] = left_node
    else:
      graph[node_number] = right_node

    if(graph[node_number] > alpha):
      alpha = graph[node_number]

    return alpha, beta
  
  elif(level == 1):
    new_alpha, new_beta = dfs(graph, alpha, beta, level+1, 2*node_number + 1)

    if(new_alpha < beta):
      beta = new_alpha

    if(alpha < beta):
      new_alpha, new_beta = dfs(graph, alpha, beta, level+1, 2*node_number + 2)

    if(new_alpha < beta):
      beta = new_alpha

    graph[node_number] = beta

    return alpha, beta
  
  else:
    new_alpha, new_beta = dfs(graph, alpha, beta, level+1, 2*node_number + 1)

    if(new_beta > alpha):
      alpha = new_beta

    if(alpha < beta):
      new_alpha, new_beta = dfs(graph, alpha, beta, level+1, 2*node_number + 2)

    if(new_beta > alpha):
      alpha = new_beta

    graph[node_number] = alpha
    return graph[0]

def make_graph(l):
  graph = []
  for i in range(7):
    graph.append(None)

  graph = graph + l
  return dfs(graph, -100000000, 100000000, 0, 0)

id = input("Enter your ID: ")
id = id.replace("0", "8")
shuffle = int(id[3])
total_score = int(id[7] + id[6]) # 59
min = int(id[4]) # 1
max = math.ceil(total_score * 1.5) # 89

l = []
shuffle_score = []
for i in range(8):
  l.append(random.randint(min, max))
  
score = make_graph(l)
print(f"Generated 8 random points between the minimum and maximum limits: {l}")
print(f"Total points to win: {total_score}")
print(f"Achieved point by applying alpha-beta pruning: {score}")
if(score >= total_score):
  print("The winner is Optimus Prime")
else:
  print("The winner is Megatron")

max_shuffle_score = 0
count = 0
for i in range(shuffle):
  random.shuffle(l)
  shuffle_score.append(make_graph(l))
  
  if(shuffle_score[i] >= total_score):
    count += 1
  if(shuffle_score[i] > max_shuffle_score):
    max_shuffle_score = shuffle_score[i]

print("After shuffle:")
print(f"List of all points values from each shuffle: {shuffle_score}")
print(f"The maximum of all shuffle: {max_shuffle_score}")
print(f"Won {count} times out of {shuffle} number of shuffles")
```
